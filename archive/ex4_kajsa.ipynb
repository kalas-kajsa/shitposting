*b) Imperfect compliance*

Just because someone is called, does not mean they will pick up. There will thus be a lot of people who were assigned treatment, but who will not receive it. They will thus be never-takers, even though it is of no fault of their own. The opposite will probably not be true; it is unlikely that people who were not assigned treatment will somehow receive the message that this is a particularly close election. So, there will be a lot of never-takers (who were assigned treatment but ended up not getting it), which will influence the average treatment effects. 

*c) Di and Zi*

Di, is our actual treatment variable, which in this dataset is *pivotal_contact*. Zi is the dummy for being assigned to a treatment, which in our case is the variable *pivotal*.

*d) ITT of regressing the indicator for voting on Zi (pivotal) + covariate check*

```python
#import packages
from statsmodels.api import add_constant
from linearmodels.iv import IV2SLS
```

```python
#Estimate LATE with a 2SLS model
df['const'] = 1

ivmod = IV2SLS(df['s11'], exog = df['const'], endog = df['pivotal_contact'], instruments = df['pivotal'])

iv_res = ivmod.fit(cov_type = 'robust')
iv_res
```

*g) perform two step regression*

```python
#First regress the treatment indicator,Di, on the instrument,Zi(usingOLS as usual).  
model = smf.ols(formula = 'pivotal_contact ~ pivotal', data = df)
results = model.fit(cov_type= 'HC0')
    
d_hat = results.fittedvalues

#regress  the  outcome  on  the  fitted  values  from  before, Dhat
model = smf.ols(formula = 's11 ~ d_hat', data = df)
results = model.fit(cov_type= 'HC0')
    
results.summary()
```
